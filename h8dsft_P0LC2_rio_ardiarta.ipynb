# SQL

```python
# 1. Kota apa saja yang masuk ke dalam cakupan wilayah distribusi?

SELECT DISTINCT store_id, store_name, store_city FROM `ftds-hacktiv8-project.phase0_ftds_003_sby.beverage_sales`

# cakupan wilayah distribusi menyeluruh ke semua store
```

```python
# 2. Berapa rentang harga minimum dan maksimum dari unit minumam yang dijual?

SELECT  MAX(unit_sold) max,
        MIN(unit_sold) min,
        AVG(unit_sold) average
FROM `ftds-hacktiv8-project.phase0_ftds_003_sby.beverage_sales`

# Jumlah harga minimum dan maksimum
```

```python
 # 3. Bandingkan total penjualan dan jumlah transaksi dari masing-masing kota pada tahun 2022?

SELECT  COUNT(unit_sold) as unit_terjual,
        COUNT(unit_retail_price) as unit_pemesanan,
        EXTRACT (YEAR FROM date) AS YEAR
FROM `ftds-hacktiv8-project.phase0_ftds_003_sby.beverage_sales`
WHERE EXTRACT (YEAR FROM date) IN (2022)
GROUP BY 3

# Jumlah transaksi pada tahun 2022
```

```python
# 4. Produk apa yang paling banyak diminati pada tahun 2022?

SELECT  unit_sold,
        product_name,
        EXTRACT (YEAR FROM date) AS YEAR,
        AVG(unit_sold) AS terjual
FROM `ftds-hacktiv8-project.phase0_ftds_003_sby.beverage_sales`
WHERE EXTRACT (YEAR FROM date) = 2022
GROUP BY product_name, unit_sold, year
ORDER BY terjual DESC

# product yang diminati di tahun 2022
```

```python
df = client.query('''
SELECT *
  FROM `ftds-hacktiv8-project.phase0_ftds_003_sby.beverage_sales`
''').to_dataframe()

df
```

```python
# perubahan jenis musim berdasarkan bulan
quarter = []

# fungsi loop for
for i in range (len(df)):

  month = df.loc[i,'date'].month
  if month in(3, 4, 5):
    quarter.append("Spring")
  elif month in (6, 7, 8):
    quarter.append("Summer")
  elif month in (8, 9, 10):
    quarter.append("Autumn")
  else:
    quarter.append("Winter")

df['quarter'] = quarter
df
```
